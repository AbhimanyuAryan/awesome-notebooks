# Send Stocks briefs by email

This notebook workflow will send you automatically stock data + latest news, by email, every day at 9:00 AM CET

Example : Tesla stock (ticker : TSLA)

## Step 1. Get the company stock data

```python
action = 'TSLA'
```

```python
import naas_drivers
import datetime as dt
```

### Define time horizon and use naas drivers to get the stock chart 

```python
start = dt.datetime.today() - dt.timedelta(days=100)
fig = naas_drivers.plot().stock([action], start, filter=False)
fig.show(config={'displayModeBar': False})
```

```python
html_file = f"{action}.html"
image_file = f"{action}.png"
css = ".main-svg {top: -100px !important;}"
naas_drivers.plot().export(fig, image_file, css)
naas_drivers.plot().export(fig, html_file,css)
```

### Import Naas and add image to assets 

```python
import naas
```

```python
url_img = naas.assets.add(image_file)
```

```python
url_html = naas.assets.add(html_file, params={"inline": True})
```

### Add news data

```python
news = []
for element in stories:
    new_formated = {
                    "title": element['title'],
                    "link": element['url'],
                    "source": element['source']['brandName'],
                    "sourceImg": element['source']['imageUrls']['large']
                  }
    news.append(new_formated)

print('News:', len(news))
pd.DataFrame.from_records(news) # use of this?
```

## Step 3. Format email 

```python
top = news[:5] # define number of news in the email ?
```

```python
import naas_drivers
import uuid

cover = naas_drivers.html().cover(url_img)
formated_news = []
for new in top:
    formated_news.append(naas_drivers.html().logo(new['sourceImg']))
    formated_news.append(naas_drivers.html().text(new['title']))
    formated_news.append(naas_drivers.html().link(new['link'], 'Read'))
formated_news
content = [
    naas_drivers.html().subtitle("👉 Analyze daily performance over time."),
    naas_drivers.html().table(formated_news, column=3, border=True),
    naas_drivers.html().text("🚀 Explore over 100 last days of stock variations. Picture are nice but dynamic chart below are way more fun  :")]
button = naas_drivers.html().button("Dynamic Chart", f"{url_html}?uid={str(uuid.uuid4())}", "white", "black")

email_res = naas_drivers.html().generate(
    f'Evolution of {action} stock', 
    cover=cover, content=content, button=button)
```

### Step 4: Send the emails

```python
naas.notifications.send("martin.donadieu@cashstory.com", f"📈 {action} situation", email_res)
naas.notifications.send("jeremy.ravenel@cashstory.com", f"📈 {action} situation", email_res)
# can we make a list to not have to repeat ? send_to = ["martin.donadieu@cashstory.com","jeremy.ravenel@cashstory.com"]
```

### Step 5: Add Scheduler

```python
naas.scheduler.add(recurrence="0 9 * * *")
```
